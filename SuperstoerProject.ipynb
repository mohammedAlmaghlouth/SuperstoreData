```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Group_112_A.csv',delimiter = ',')
display(df)
df.info()
display(df.describe())
```

```python
#fixing inconsistency
#as we can see in the proximity column, some numbers are given with the unit and we need to remove it
df['Proximity'] = df['Proximity'].apply(lambda x: int(x.replace('mts','')))
```

```python
selected_cols = df.select_dtypes(exclude='object').columns.tolist()
df[selected_cols] = df[selected_cols].applymap(lambda x: np.abs(x)) #just to make sure that no number is negative
```

```python
#removing outliers
selected_cols = df.select_dtypes(exclude='object').columns.tolist()
for col in selected_cols:
    col_array = df[col].values
    Q3, Q1 = np.percentile(col_array, [75 ,25])
    IQR = Q3 - Q1
    upperWhisker = Q3 + 1.5 * IQR
    lowerWhisker = Q1 - 1.5 * IQR
    print(lowerWhisker,upperWhisker) #to check that no value exceed the upper or lower whisker
    mask = (col_array > upperWhisker) | (col_array < lowerWhisker) #outliers lie above and below these boundries
    outliers = col_array[mask]
    selected_rows = df[col].isin(outliers)
    index = df[selected_rows].index
    df.drop(index , inplace=True) #to get rid of outliers
df.sample(5)
#note: although outliers shall only be removed from Sale-Price and Purchase-Price we did it for all the numeric values
# and so, we looped over the whole numeric columns
```

```python
#handling missing values
null_columns=df.columns[df.isna().any()] # four fields have NaN values 
print(null_columns) 
```

```python
ndf = df[(df['Type']!='open land') & (df['Type']!='land')]
null_columns=ndf.columns[ndf.isna().any()]
print(null_columns)

#ndf
#since for all types applicable for renovation, there exist a value 
#so NaN values for (open land) and (land) should be filled with zeros since they are not applicable for renovation

df['Renovate'].fillna(value=0,inplace=True)
null_columns=df.columns[df.isna().any()]
print(null_columns) #we eliminated one field and remain 3
```

```python
ndf = df[(df['Type']=='open land') | (df['Type']=='land')] #in the same way, if the type was (open land) or (land) then 
#no missing values will exist
null_columns=ndf.columns[ndf.isna().any()]
print(null_columns)

#so we have to fill the other types i.e.(villa,duplex,flat) with a value. 
#if not applicable we fill with the string 'NV' which indicates that there is no value 

df['Contour'].fillna(value='NV',inplace=True)
null_columns=df.columns[df.isna().any()]
print(null_columns) #we eliminated two fields and remain 2
```

```python
ndf = df[(df['Type']=='duplex') | (df['Type']=='flat') | (df['Type']=='villa')]
null_columns=ndf.columns[ndf.isna().any()]
print(null_columns)
#null values exist only for non-applicable types so we fill them with NV as before

df['Garage'].fillna(value='NV',inplace=True)
null_columns=df.columns[df.isna().any()]
print(null_columns) #we eliminated three fields and remain 1
```

```python
ndf = df[(df['Type']=='villa')]
null_columns=ndf.columns[ndf.isna().any()]
print(null_columns)

#same as before

df['Swimming'].fillna(value='NV',inplace=True)
null_columns=df.columns[df.isna().any()]
print(null_columns) #we eliminated all of missing values and we filled them
```

## Task-4

```python
#we will use PCA 
dflist = ['Sale-Price','Purchase-Price']

df['Profit Percentage'] = ((df[dflist[0]]-df[dflist[1]])/df[dflist[1]])*100

df['high/low']= df['Profit Percentage'].apply(lambda x: 'high' if x>=400 else 'low' if x<=100 else 'medium' )


num_columns = df.select_dtypes(exclude='object').columns
pca_input_columns=num_columns.drop(['Profit Percentage','Sale-Price'])
print(pca_input_columns)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df[pca_input_columns])
df['pca1'] = principalComponents[:,0]
df['pca2'] = principalComponents[:,1]

display(df)


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.relplot(x='pca1',y='pca2',
            hue='high/low', palette=['r','b','g'],
            kind='scatter',alpha=0.75,
            height=5, aspect=1,
            data=df)
plt.title('PCA vs high/low', fontsize = 20)
plt.show()

df.drop(['pca1','pca2'], axis=1, inplace=True)
#the graph shows a good (but not perfect) seperation of the three categories high,medium,low
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

df = pd.read_csv('Group_112_B.csv', delimiter=',')


num_columns = df.select_dtypes(exclude='object').columns


Xo = df[num_columns].iloc[:, 1:].values
y = df[num_columns].iloc[:, 0].values


Xo_train, Xo_test, y_train, y_test = train_test_split(
    Xo, y, test_size=0.3, random_state=42)


reg = LinearRegression().fit(Xo_train, y_train)

best_beta = np.round(reg.coef_, 2)
best_beta_0 = np.round(reg.intercept_, 2)


print(f'The best values for the estimates are :',
      best_beta_0, best_beta.tolist())
```

## Task-8

```python
from sklearn.linear_model import Ridge
regr = Ridge(alpha=500)
regr.fit(Xo_train, y_train)
best_beta = np.round(regr.coef_, 2)
best_beta_0 = np.round(regr.intercept_, 2)
print(f'The best values for the estimates are :',
      best_beta_0, best_beta.tolist())
```
